```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Leer el CSV
df = pd.read_csv('/home/StressLevelDataset.csv')

# DIAGRAMA DE BARRAS
plt.figure(figsize=(12, 6))

# Opción 1: Barras agrupadas por nivel de ansiedad
anxiety_counts = df['anxiety_level'].value_counts().sort_index()

plt.bar(anxiety_counts.index, anxiety_counts.values, color='red', alpha=0.7)
plt.title('Distribución de Niveles de Ansiedad')
plt.xlabel('Nivel de Ansiedad')
plt.ylabel('Cantidad de Personas')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Leer el CSV
df = pd.read_csv('/home/heart_disease_dataset.csv')

# DIAGRAMA DE BARRAS AGRUPADAS POR SEXO
plt.figure(figsize=(14, 6))

# Separar datos por sexo (asumiendo que la columna se llama 'sex')
# Ajusta el nombre si es diferente (podría ser 'gender', 'sexo', etc.)
hombres = df[df['sex'] == 1]  # Usualmente 1 = hombre, 0 = mujer
mujeres = df[df['sex'] == 0]  # Ajusta estos valores según tu dataset

# Contar edades por sexo
age_counts_hombres = hombres['age'].value_counts().sort_index()
age_counts_mujeres = mujeres['age'].value_counts().sort_index()

# Obtener todas las edades únicas para alinear las barras
all_ages = sorted(df['age'].unique())

# Rellenar con 0 las edades que no aparecen en cada grupo
hombres_counts = [age_counts_hombres.get(age, 0) for age in all_ages]
mujeres_counts = [age_counts_mujeres.get(age, 0) for age in all_ages]

# Configurar las posiciones de las barras
x = range(len(all_ages))
width = 0.35  # Ancho de las barras

# Crear barras agrupadas
plt.bar([i - width/2 for i in x], hombres_counts, width,
        label='Hombres', color='blue', alpha=1)
plt.bar([i + width/2 for i in x], mujeres_counts, width,
        label='Mujeres', color='fuchsia', alpha=1)

# Configurar la gráfica
plt.title('Distribución de Edades por Sexo - Enfermedades Cardíacas')
plt.xlabel('Edad')
plt.ylabel('Cantidad de Personas')
plt.xticks(x, all_ages, rotation=45)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Leer el CSV
df = pd.read_csv('/home/wine_quality_merged.csv')

# DIAGRAMA LINEAL
plt.figure(figsize=(12, 8))

# Obtener la distribución de pH y ordenarla
ph_counts = df['pH'].value_counts().sort_index()

# Crear el gráfico lineal
plt.plot(ph_counts.index, ph_counts.values,
         color='lightGreen', linewidth=2.5, marker='o',
         markersize=6, markerfacecolor='green',
         markeredgecolor='darkGreen', markeredgewidth=1.5)

# Personalización del gráfico
plt.title('Distribución de pH en Vinos', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Nivel de pH', fontsize=12, fontweight='bold')
plt.ylabel('Cantidad de Vinos', fontsize=12, fontweight='bold')

# Mejorar la apariencia
plt.grid(True, alpha=0.3, linestyle='--')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
